### ノック13 : データに揺れがあるまま集計してみよう

```python
uriage_data['purchase_date'] = pd.to_datetime(uriage_data['purchase_date'])
uriage_data['purchase_month'] = uriage_data['purchase_date'].dt.strftime('%Y%m')

res = uriage_data.pivot_table(index='purchase_month', columns='item_name', aggfunc='size', fill_value=0)

res
```

```python
res = uriage_data.pivot_table(index='purchase_month', columns='item_name', values='item_price', aggfunc='sum', fill_value=0)

res
```

### ノック14 : 商品名の揺れを補正しよう

```python
len(pd.unique(uriage_data['item_name']))
```

```python
# 名寄せしていく

# アルファベットを大文字にする
uriage_data['item_name'] = uriage_data['item_name'].str.upper()

# 全角スペースを消す
uriage_data['item_name'] = uriage_data['item_name'].str.replace('　', '')

# 半角スペースを消す
uriage_data['item_name'] = uriage_data['item_name'].str.replace(' ', '')

uriage_data.sort_values(by=['item_name'], ascending=True)
```

### ノック15 : 金額欠損値を補完してみよう

```python
# 金額に欠損があるかの確認
uriage_data.isnull().any(axis=0)
```

```python
flg_is_null = uriage_data['item_price'].isnull()

for target in list(uriage_data.loc[flg_is_null, 'item_name'].unique()):

    price = uriage_data.loc[(~flg_is_null) & (uriage_data['item_name'] == target), 'item_price'].max()
    
    uriage_data.loc[(flg_is_null) & (uriage_data['item_name'] == target), 'item_price'] = price
    
uriage_data.isnull().any(axis=0)
```

### ノック16 : 顧客名の揺れを補正しよう

```python
kokyaku_data['顧客名']
```

```python
uriage_data['customer_name']
```

```python
# 名寄せしていく
# .. 全角スペースを消す
kokyaku_data['顧客名'] = kokyaku_data['顧客名'].str.replace('　', '')

# .. 半角スペースを消す
kokyaku_data['顧客名'] = kokyaku_data['顧客名'].str.replace(' ', '')

kokyaku_data['顧客名']
```

### ノック17 : 日付の揺れを補正しよう

```python
flg_is_serial = kokyaku_data['登録日'].astype('str').str.isdigit()

# 数字認識されているセル数
flg_is_serial.sum()
```

```python
from_serial = pd.to_timedelta(kokyaku_data.loc[flg_is_serial, '登録日'].astype('float') - 2, unit='D') + pd.to_datetime('1900/1/1')

from_serial
```

```python
from_string = pd.to_datetime(kokyaku_data.loc[~flg_is_serial, '登録日'])

from_string
```

```python
kokyaku_data['登録日'] = pd.concat([from_serial, from_string])

kokyaku_data
```

```python
kokyaku_data['登録年月'] = kokyaku_data['登録日'].dt.strftime('%Y%m')

rslt = kokyaku_data.groupby('登録年月').count()['顧客名']

print(rslt)
print(len(kokyaku_data))
```

### ノック18 : 顧客名をキーに2つのデータを結合（join）してみよう

```python
# 売り上げデータと顧客データを　customer_name == 顧客名 で left join 
join_data = pd.merge(uriage_data, kokyaku_data, left_on='customer_name', right_on='顧客名', how='left')

join_data = join_data.drop('customer_name', axis=1)

join_data
```

### ノック19 : クレンジングしたデータをダンプしよう

```python
join_data.to_csv('sample_20201021/chapter_002/dump_data_20230109.csv')
```
